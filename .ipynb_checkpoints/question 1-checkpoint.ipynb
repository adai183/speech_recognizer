```python
from glob import glob
import numpy as np
import _pickle as pickle
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style(style='white')

def plot_losses(model_paths):
    # obtain the paths for the saved model history
    
    model_names = [item[8:-7] for item in model_paths]
    # extract the loss history for each model
    valid_loss = [pickle.load( open( i, "rb" ) )['val_loss'] for i in model_paths]
    train_loss = [pickle.load( open( i, "rb" ) )['loss'] for i in model_paths]
    # save the number of epochs used to train each model
    num_epochs = [len(valid_loss[i]) for i in range(len(valid_loss))]

    fig = plt.figure(figsize=(16,5))

    # plot the training loss vs. epoch for each model
    ax1 = fig.add_subplot(121)
    for i in range(len(model_paths)):
        ax1.plot(np.linspace(1, num_epochs[i], num_epochs[i]), 
                train_loss[i], label=model_names[i])
    # clean up the plot
    ax1.legend()  
    ax1.set_xlim([1, max(num_epochs)])
    plt.xlabel('Epoch')
    plt.ylabel('Training Loss')

    # plot the validation loss vs. epoch for each model
    ax2 = fig.add_subplot(122)
    for i in range(len(model_paths)):
        ax2.plot(np.linspace(1, num_epochs[i], num_epochs[i]), 
                valid_loss[i], label=model_names[i])
    # clean up the plot
    ax2.legend()  
    ax2.set_xlim([1, max(num_epochs)])
    plt.xlabel('Epoch')
    plt.ylabel('Validation Loss')
    plt.show()
    
all_pickles = sorted(glob("results/*.pickle"))
# do not plot the end model for now
plot_losses(all_pickles[:6])
```

### Model 0: Simple RNN

### Model 2 : CNN + RNN + TimeDistributed Dense


Model 2 add a convolution as first layers. This way our Model is able to extract more complex features and we also aproximately double the number of parameters. Model 2 is the performance winner in terms of training loss but has a problem with overfitting and only gets second place on the validation set. This means in order to improve performance we need to apply regularization.

### Model 3:  Deeper RNN + TimeDistributed Dense

Is the best model because it performs best on the validation set. This is the case because it applys regularization through Batch Normalization after every RNN layer. It also has less parameters than Model 2 and trains faster.

**Conclusion**: We should experiment with even deeper architectures!

### Model 4: Bidirectional RNN + TimeDistributed Dense

Model 4 experiments with a bidirectional RNN layer. So now our model is able to better understand the context in both directions. This gives us results close to the best but slightly worse. Which is impressive because it only uses one RNN layer.

**Conclusion:** We should experiment with deeper architectures and use more than one bidirectional RNN layer.
